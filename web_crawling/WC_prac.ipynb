# 웹 크롤링

웹 크롤링 기법  
1. 웹 문서를 텍스트로 읽어서 가져오는 법 (beatifulsoup4)
2. 웹 브라우저를 매크로 시켜서 가져오는 법 (selenium)

```python
from bs4 import BeautifulSoup as bs
```

```python
import re, usecsv, requests
import urllib.request as ur
```

```python
url  = 'http://quotes.toscrape.com/'
html = ur.urlopen(url)
```

```python
# bs(html.read(), 'html.parser') : 파싱한 것
soup = bs(html.read(),'html.parser')
print(soup)
```

```python
#find_all(tag) : 알맞는 태그를 찾아주는 함수
soup.find_all('span')
```

```python
#<span>by <small class="author" itemprop="author">Albert Einstein</small>
#<a href="/author/Albert-Einstein">(about)</a>
#</span>

quote = soup.find_all('span')
print(quote[1].text) #text : 태그 사이에 있는 내용만

for i in quote:
    print(i.text)
```

```python
quote = soup.find_all('div')
print(quote)
```

```python
quote2 = soup.find_all('span', {'class' : 'text'})
print(quote2[0].text)
```

```python
url  = 'https://news.daum.net/'
soup = bs(ur.urlopen(url).read(), 'html.parser')
headline = soup.find_all('a', {'class' : 'link_txt'})
for i in headline:
    print(i.text)
```
